```python
import pandas as pd

data={'SubjectName':['s1','s2','s3','s1','s4','s3','s2','s1','s2','s4','s1'],
      'Target':[1,0,1,1,1,0,0,1,1,1,0]}

df = pd.DataFrame(data)

df
```

```python
mean = df.groupby(df.SubjectName)["Target"].mean()
mean
```

```python
val = df.SubjectName.map(mean)
val
```

```python
df["Column1"] =  val
```

```python
df
```

```python
# 2nd method with apply function
df["Column2"] = df.SubjectName.apply(lambda i: mean[i])
```

# New Data

```python
df.drop(columns=["SubjectName","Column2"])
```

# Apply with category_encoders lib

```python
df.drop(columns=["Column1","Column2"],inplace=True)
```

```python
df
```

```python
from category_encoders import TargetEncoder
```

```python
te = TargetEncoder(min_samples_leaf=2)
```

```python
x = df.SubjectName
y = df.Target
```

```python
te.fit_transform(x,y)
```

# Change min_samples

```python
te = TargetEncoder(min_samples_leaf=-10)
te.fit_transform(x,y)
```
